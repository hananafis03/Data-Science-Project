```python
#Heatmap showing correlation between dataframes
new = data.drop(columns=['Education', 'EnvironmentSatisfaction','JobLevel','PercentSalaryHike', 'MonthlyIncome','WorkLifeBalance','JobInvolvement', 'PerformanceRating','JobSatisfaction', 'TrainingTimesLastYear','StockOptionLevel','EmployeeID','DistanceFromHome', 'NumCompaniesWorked', 'EmployeeCount', 'StandardHours', 'BusinessTravel', 'Department', 'EducationField', 'Gender', 'JobRole', 'MaritalStatus','Over18'], axis = 1).set_index('Attrition')
new.head()
plt.figure(figsize = (10,3))
sns.heatmap(new.corr(), annot = True, linewidth = 1)
plt.title("Heatmap showing correlation between elements in the dataset")
plt.show()
```

```python
#Histogram of "Age" feature
plt.figure(figsize = (10,5))
median = round(data['Age'].median())
mean = round(data['Age'].mean())

sns.histplot(data=data, x='Age', color='blue', edgecolor='linen', alpha =0.5, bins=5)
plt.title("Histogram of Employee Age")
plt.xlabel('Age')
plt.ylabel('Count')
plt.vlines(mean, ymin = 0, ymax = 1600, color ='red', label="Mean")
plt.vlines(median, ymin = 0, ymax = 1600, color ='green', label="Median")
plt.legend()
plt.show()
```

```python
#Count plot of "Job Level" feature
plt.figure(figsize = (10,5))

sns.countplot( x =data['JobSatisfaction'], color = 'yellow')
plt.title("Count Plot of Job Satisfaction")
plt.xlabel('Job Satisfaction')
plt.ylabel('Count')
plt.show
```

```python
#Count plot of "Education Level" feature
plt.figure(figsize = (10,5))

sns.countplot(data=data, x ='Education', color = 'green')
plt.title("Count Plot of Education")
plt.xlabel('Education')
plt.ylabel('Count')
plt.show
```

```python
#Count plot of "Education Field" feature
plt.figure(figsize = (10,5))

sns.countplot(data=data, x ='EducationField', color = 'red')
plt.title("Count Plot of Education Field")
plt.xlabel('EDUCATION FIELD')
plt.ylabel('Count')
plt.show
```

```python
#Count plot of "Job Role" feature
plt.figure(figsize = (20,5))

sns.countplot(data=data, x ='JobRole', color = 'green')
plt.title("Count Plot of Job Role")
plt.xlabel('JOB ROLE')
plt.ylabel('COUNT')
plt.show
```

```python
#Pie chart of education field
plt.figure(figsize = (10,6))
sums = data.groupby(data['EducationField'])['Education'].sum()
plt.pie(sums, labels=sums.index, autopct='%.2f')
plt.title("Education Field Chart")
plt.show()
```

```python
#Pie chart of job role
plt.figure(figsize = (10,6))
sums = data.groupby(data['JobRole'])['JobLevel'].sum()
plt.pie(sums, labels=sums.index, autopct='%.2f')
plt.title("Job Role Chart")
plt.show()
```
